# Agg_transactions

Transform the data into suitable format

```python
import pandas as pd
import json
import os
```

```python
path1 = "/content/Pulse/data/aggregated/transaction/country/india/state"
agg_trans_list = os.listdir(path1)
# agg_trans_list
clm = {'State':[],'Year':[],'Quater':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}
for i in agg_trans_list:
  p_i = path1+"/"+i
  agg_yr = os.listdir(p_i)
  #print(agg_yr)
  for j in agg_yr:
    p_j = p_i+"/"+j
    agg_yr_list = os.listdir(p_j)
    #print(agg_yr_list)
    for k in agg_yr_list:
      p_k = p_j+"/"+k
      data = open(p_k,"r")
      A = json.load(data)
      for z in A["data"]["transactionData"]:
        Name = z["name"]
        count = z["paymentInstruments"][0]["count"]
        amount = z["paymentInstruments"][0]["amount"]
        clm["Transaction_type"].append(Name)
        clm["Transaction_count"].append(count)
        clm["Transaction_amount"].append(amount)
        clm["State"].append(i)
        clm["Year"].append(j)
        clm["Quater"].append(int(k.strip('.json')))
# create a DataFrame
df_agg = pd.DataFrame(clm)
```

```python
df_agg.to_csv('agg_transaction')
```

```python
df_agg.head()
```

```python
# insert the dataframe into sql lite base
import sqlite3
con = sqlite3.connect('pulse.db')
c = con.cursor()
```

```python
# create a table in the database
c.execute(''' CREATE TABLE IF NOT EXISTS TRANSACTION_DATA(
  STATE TEXT,
  YEAR TEXT,
  QUATER TEXT,
  TRANSACTION_TYPE TEXT,
  TRANSACTION_COUNT TEXT,
  TRANSACTION_AMOUNT TEXT
)''')
```

```python
# Insert all the data into table
c.executemany("INSERT INTO TRANSACTION_DATA VALUES(?,?,?,?,?,?)",
              zip(clm["State"],clm["Year"],clm["Quater"],clm["Transaction_type"],clm["Transaction_count"],clm["Transaction_amount"]))
```

```python
# show the data in table format
x = c.execute("SELECT * FROM TRANSACTION_DATA")
x = c.fetchall()
for row in x:
  print(row)
```

```python
# commit the changes
con.commit()
con.close()
```

# Map_insurance

```python
path5 = "/content/Pulse/data/map/insurance/hover/country/india/state"
map_insur = os.listdir(path5)
clm = {'State':[],'district':[],'Year':[],'Quarter':[],'Metric_type':[],'Metric_count':[],'Metric_amount':[]}
for i in map_insur:
  m_i = path5+'/'+i
  map_yr = os.listdir(m_i)
  for j in map_yr:
    m_j = m_i+'/'+j
    map_yr_list = os.listdir(m_j)
    for k in map_yr_list:
      m_k = m_j+'/'+k
      data = open(m_k,'r')
      B = json.load(data)

      #Check if 'hoverDataList' exists and is non-empty
      if 'hoverDataList' not in B['data'] or not B['data']['hoverDataList']:
        continue

      for item in B['data']['hoverDataList']:
        # Extract data safely
        district_name = item.get('name','unknown')  #This gets the 'name' key (district name),
        # if 'name' is missing, it uses 'unknown' to avoid crashing.

      clm['district'].append(district_name)
      clm['State'].append(i)
      clm['Year'].append(j)
      clm['Quarter'].append(int(k.strip('.json')))
      for metric in B['data']['hoverDataList'][0]['metric']:
        clm['Metric_type'].append(metric['type'])
        clm['Metric_count'].append(metric['count'])
        clm['Metric_amount'].append(metric['amount'])
# create a DataFrame
map_ins = pd.DataFrame(clm)
```

```python
# downloading the file into csv formate
map_ins.to_csv('map_insurance')
```

# Map --> user

```python
path6 = '/content/Pulse/data/map/user/hover/country/india/state'
map_us = os.listdir(path6)
clm = {'State': [], 'Year': [], 'Quarter': [], 'No of users': [], 'AppOpen':[], 'District':[] }
for i in map_us:
  m_i = path6+'/'+i
  map_yr = os.listdir(m_i)
  for j in map_yr:
    m_j = m_i+'/'+j
    map_yr_list = os.listdir(m_j)
    #print(map_yr_list)
    for k in map_yr_list:
      m_k = m_j+'/'+k
      data = open(m_k,'r')
      D = json.load(data)
      clm['State'].append(i)
      clm['Year'].append(j)
      clm['Quarter'].append(int(k.strip('.json')))
      d_i = D['data']['hoverData']
      for dic,sta in d_i.items():
        dist = dic
        users = (sta['registeredUsers'])
        app = (sta['appOpens'])
      clm['District'].append(dist)
      clm['No of users'].append(users)
      clm['AppOpen'].append(app)
# create a DataFrame
map_use = pd.DataFrame(clm)
```

```python
# downloading the file into csv formate
map_use.to_csv('map_user')
```
